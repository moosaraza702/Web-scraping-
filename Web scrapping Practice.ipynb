```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
 results = soup.find_all('div', attrs={'class':'blogText'})
```

```python
para = soup.find_all('p')
para[0]
```

```python
record = []
count = 0
count1 = 1
for result in soup:
    h = soup.find_all('strong')
    heading  = h[count].contents[0]
    p = soup.find_all('p')
    para=p[count1].contents[1]
    record.append((heading,para))
    count = count +1
    count1 = count1 + 1
    if count == 5:
        break
    else:
        continue
    if count1 == 5:
        break
    else:
        continue
```

```python
record
```

```python
import pandas as pd
data = pd.DataFrame(record, columns = ['heading', 'Paragraph'])
```

```python
data.head()
```

```python
data.to_csv('Indian media about Imran Khan',index=False, encoding='utf-8')
```
